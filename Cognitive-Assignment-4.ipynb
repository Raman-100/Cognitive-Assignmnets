```python
import numpy as np

x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
most_freq_value = np.bincount(x).argmax()
indices = np.where(x == most_freq_value)[0]

print(f'Most frequent value: {most_freq_value}')
print(f'Indices: {indices}')

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_freq_value_y = np.bincount(y).argmax()
indices_y = np.where(y == most_freq_value_y)[0]

print(f'Most frequent value: {most_freq_value_y}')
print(f'Indices: {indices_y}')
```

```python
import numpy as np

your_name = np.linspace(10, 100, 25)
dimensions = your_name.ndim
shape = your_name.shape
total_ele = your_name.size
data_type = your_name.dtype
total_bytes = your_name.nbytes

print(f'Dimensions: {dimensions}')
print(f'Shape: {shape}')
print(f'Total elements: {total_ele}')
print(f'Data type: {data_type}')
print(f'Total bytes: {total_bytes} bytes')

transpose_array = your_name.reshape(25, 1)
print(f'Transpose using reshape(): \n{transpose_array}')

transpose_array_T = your_name.T
print(f'Transpose using T attribute: \n{transpose_array_T}')
```

```python
import numpy as np

ucs420_your_name = np.array([[10, 20, 30, 40], [50, 60, 70, 80], [90, 15, 20, 35]])
mean_value = np.mean(ucs420_your_name)
median_value = np.median(ucs420_your_name)
max_value = np.max(ucs420_your_name)
min_value = np.min(ucs420_your_name)
unique_elements = np.unique(ucs420_your_name)

print(f'Mean: {mean_value}')
print(f'Median: {median_value}')
print(f'Max: {max_value}')
print(f'Min: {min_value}')
print(f'Unique elements: {unique_elements}')

reshaped_ucs420_your_name = ucs420_your_name.reshape(4, 3)
print(f'Reshaped array (4x3): \n{reshaped_ucs420_your_name}')

resized_ucs420_your_name = np.resize(ucs420_your_name, (2, 3))
print(f'Resized array (2x3): \n{resized_ucs420_your_name}')
```
