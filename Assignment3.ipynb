```python
import pandas as pd

data={
    "Employee_Id": [101,102,103,104,105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29,34,41,28,38],
    'Salary': [50000,70000,65000,55000,60000],
    'Years_of_Experience':[4,8,10,3,12],
    'Joining Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000,7000,6000,4500,5000],
    'Rating': [4.5,4.0,3.8,4.7,3.5],
}

df = pd.DataFrame(data)
df.to_csv('employees.csv', index=False)

#from google.colab import files
#files.download('employees.csv')

df=pd.read_csv('employees.csv')
print(df,'\n')
print(df.shape,'\n')
summary=df.info
print(summary,'\n')
statistics = df.describe()
print(statistics,'\n')
df.head()
df.tail(3)
print("Average salary of the employees is: ", df['Salary'].mean())
print("Total bonus paid to employees is : ", df['Bonus'].sum())
print("Youngest employee's age is: ", df['Age'].min())
print("Highest performance rating is: ", df['Rating'].max())
print('\n')
sort_salary = df.sort_values(by='Salary')
print('Sorting dataframe by the salary column in ascending order: \n',sort_salary)
print('\n')
sort_salary2 = df.sort_values(by='Salary', ascending=False)
print("Salary in descing order: \n", sort_salary2)
print('\n')
def catogerize_performance(rating):
  if rating>=4.5:
    return 'Excfellent'

  elif rating>=4.0 and rating<4.5:
    return 'Good'

  else:
    return 'Average'

df['Performance Rating'] = df['Rating'].apply(catogerize_performance)
print(df)
df = df.rename(columns={'Employee_Id': 'ID'})
print('After renaming: \n')
print(df)
filter = df[(df['Years_of_Experience']>5) & (df['Department'] == 'IT')]
print(filter)

modified_csv_path = "modified_employees.csv"
df.to_csv(modified_csv_path, index=False)

print(f"Modified DataFrame saved to {modified_csv_path}")
```
