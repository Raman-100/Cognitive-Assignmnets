```python
marks = (45,89.5,76,45.5,89,92,58,45)
#highest score and its index
print('Maximum marks is: ',max(marks))
maxmarks = marks.index(max(marks))
print('Maximum marks index is: ',maxmarks)
#lowest and count it
print('Minimum score is: ', min(marks))
freq = marks.count(min(marks))
print('Number of times the lowest score appear is: ', freq)
#reverse
score = list(marks)
print('Reverse is: ', score[::-1])
#check if present
if 76 in marks:
    print('Yes 76 is present')
else:
    print('No, not present')
```

```python
# Generate a list of 100 random numbers between 100 and 900
import random
Random = [random.randint(100,900) for _ in range(100)]
listeven = []
listodd = []
listprime = []
print(Random)
for x in Random:
    if x%2==0:
        listeven.append(x)
    else:
        listodd.append(x)

for x in Random:
    for i in range(2,int(x**0.5)+1):
        if x%i==0:
            listprime.append(x)

print('Random even integers are: ', listeven)
print('Random odd integers are: ', listodd)
print('Random prime integes are: ', listprime)
```

```python
A = {34,56,78,90}
B = {78,45,90,23}
print('Unique scores are: ', A.union(B))
print('Scores that are common to both teams: ', A.intersection(B))
print('Scores that are exclusive to each team: ', A.symmetric_difference(B))
n=int(input("Enter the score to delete: "))
if(n in A):
  A.remove(n)
  print("Itemm has been removed", A)
else:
  print("Entered score has not been scored by the Team A")
```
