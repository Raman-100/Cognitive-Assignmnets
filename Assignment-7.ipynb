```python
# 1st

import numpy as np
import pandas as pd

# 1. Initialize Random Seed
np.random.seed(42)

# 2. Generate Sales Data
# Sales values between 1000 and 5000 for 12 months across 4 categories
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# 3. Convert to DataFrame
# Define months and column names
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

# Create DataFrame
df = pd.DataFrame(sales_data, columns=columns, index=months)

# Display the DataFrame
print(df)

# 1. Display the first 5 rows and summary statistics
print(df.head())
print(df.describe())

# 2. Calculate total sales per category and total sales per month
total_sales_per_category = df.sum(axis=0)  # Column-wise sum
total_sales_per_month = df.sum(axis=1)    # Row-wise sum
print("Total Sales Per Category:\n", total_sales_per_category)
print("Total Sales Per Month:\n", total_sales_per_month)

# 3. Calculate average sales growth between consecutive months for each category
average_growth = df.pct_change().mean()
print("Average Sales Growth by Category:\n", average_growth)

# 4. Add columns for Total Sales and Growth Rate
df['Total Sales'] = total_sales_per_month
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
print(df)

# 5. Apply a randomized discount
roll_number = 102317079  # Example roll number
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9  # Apply 10% discount
else:
    df['Clothing'] *= 0.85  # Apply 15% discount
print("After Discount:\n", df)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for category in df.columns[:-2]:  # Exclude 'Total Sales' and 'Growth Rate'
    plt.plot(df.index, df[category], marker='o', label=category)

plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Category')
plt.grid(True)
plt.show()

import seaborn as sns

# Melt the DataFrame for easier plotting with Seaborn
df_melted = df.reset_index().melt(id_vars='index', value_vars=columns, var_name='Category', value_name='Sales')

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Sales', data=df_melted)
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# 2nd

import numpy as np

# Define the array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Find element-wise absolute value
abs_array = np.abs(array)
print("Element-wise absolute value:\n", abs_array)

# ii. Find the 25th, 50th, and 75th percentile
# Flattened array
flattened = array.flatten()
percentiles_flattened = {
    '25th': np.percentile(flattened, 25),
    '50th': np.percentile(flattened, 50),
    '75th': np.percentile(flattened, 75)
}
print("Percentiles of flattened array:", percentiles_flattened)

# Percentiles for each column
percentiles_columns = {
    '25th': np.percentile(array, 25, axis=0),
    '50th': np.percentile(array, 50, axis=0),
    '75th': np.percentile(array, 75, axis=0)
}
print("Percentiles for each column:\n", percentiles_columns)

# Percentiles for each row
percentiles_rows = {
    '25th': np.percentile(array, 25, axis=1),
    '50th': np.percentile(array, 50, axis=1),
    '75th': np.percentile(array, 75, axis=1)
}
print("Percentiles for each row:\n", percentiles_rows)

# iii. Mean, Median, and Standard Deviation
mean_flattened = np.mean(flattened)
median_flattened = np.median(flattened)
std_flattened = np.std(flattened)
print("Flattened array - Mean:", mean_flattened, "Median:", median_flattened, "Std Dev:", std_flattened)

#Column-wise stats
mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_columns = np.std(array, axis=0)
print("Column-wise - Mean:", mean_columns, "Median:", median_columns, "Std Dev:", std_columns)

#Row-wise stats
mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)
print("Row-wise - Mean:", mean_rows, "Median:", median_rows, "Std Dev:", std_rows)
```

```python
# 3rd

import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("Original array: ", a)
print("Floor values: ", floor_values)
print("Ceiling values: ", ceiling_values)
print("Truncated values: ", truncated_values)
print("Rounded values: ", rounded_values)
```

```python
# 4th

def swap_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp

my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)

index1, index2 = 1, 3

swap_elements(my_list, index1, index2)
print("List After Swapping:", my_list)
```

```python
# 5th


def swap_elements_in_set(input_set, index1, index2):
    temp_list = list(input_set)
    temp_list[index1], temp_list[index2] = temp_list[index2], temp_list[index1]
    return set(temp_list)

my_set = {10, 20, 30, 40, 50}
print("Original Set:", my_set)

index1, index2 = 0, 3
swapped_set = swap_elements_in_set(my_set, index1, index2)
print("Set After Swapping:", swapped_set)
```
